```python
def urso(nome_arquivo):

    df = pd.read_csv(nome_arquivo,encoding="ANSI", engine='python', sep = ";", header = 0, dtype= {'Code': str})
    df['code'], df['nome'] = df.iloc[:,1].str.split(':').str
    df.drop([0,2,4,5,6], inplace = True)
    cols = [1,96]
    df.drop(df.columns[cols], inplace = True, axis = 1)
    l =['ano']
    for i in range (0,len(df.columns)-3): 
            if i < (len(df.columns)-3)/2:
                l.append("H_")
            else: 
                l.append("M_")
    l.append("")
    l.append("nome")
    df.loc[1] = l


    df.index = df ['code']
    df.drop(['code'], inplace = True,axis=1)
    df
    lista =['ano']
    for i in range (1,len(df.columns)-1): 
            lista.append(str(df.iloc[0,i]) + str(df.iloc[1,i]))
    lista.append('nome')
    df.columns = lista
    df.drop(df.index[[0,1]], inplace = True)

    anos2 = df.ano.unique()
    anos2 = np.delete (anos2,1)

    for i in range(int(len(df.index))):
        df.iloc[i,0] = anos2[0]

    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]


    l_index = []

    for i in df.index:
        if len(i) > 4: 
            l_index.append(i[3:])
        else: 
            l_index.append(i)

    df.index = l_index

    df

    cols1 = cols[2:(int(len(cols[2:])/2+2))]

    df_h = df[cols1]
    df_h = df_h.apply(pd.to_numeric)
    df_h

    cols2 = cols[(int(len(cols[2:])/2+2)):]
    df_m = df[cols2]
    df_m = df_m.apply(pd.to_numeric)

    df_m.columns = df_h.columns 

    df_add = df_m.add(df_h, fill_value=0)
    df_add['ano'] = df ['ano']

    df_soma = pd.DataFrame()
    k=0
    for i in range (0,(len(df_add.columns)//5)):
        df_soma[i] = df_add.iloc[:, k:k+5].sum(axis = 1)
        k+=5
    df_soma[i+1] = df_add.loc[:,'H_55 e mais anos']
    df_soma['ano'] = df_add.loc[:,'ano']
    l = ('10_14', '15_19', '20_24', '25_29', '30_34','35_39', '40_44', '45_49', '50_54', '55_59', 'ano')
    df_soma.columns = l 

    l_mun = []
    for i in df_soma.index:
        if len(i) == 4: 
            l_mun.append(i)

    myset = set(l_mun)
    l_mun = list(myset)
    df_soma = df_soma.loc[myset]
    df_soma.index.name = 'code'
    df_soma
    return (df_soma)
```
